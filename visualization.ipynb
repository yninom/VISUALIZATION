1. Meganoticias sobre el juicio pol√≠tico de Trump

<img src="trump.jpg" width="500">

```python
import matplotlib.pyplot as plt

# Datos
x = ["Enjuiciado pero no destituido","Abuso de poder", "Obstruccion al congreso"]
y = [41, 53, 48]

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(6, 4))

# Dibujar gr√°fico de barras horizontales
ax.barh(x, y)

# Configuraci√≥n de la gr√°fica
ax.set_title("Trump Juicio Pol√≠tico")
ax.tick_params(axis='both', which='both', length=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_xticks([])

ax.barh(x, y, color=['blue'])


# Agregar valores de y en cada barra
for i in range(len(x)):
    ax.text(y[i], i, str(y[i])+'%', ha='left', va='center')

# Mostrar la gr√°fica
plt.show()
```
